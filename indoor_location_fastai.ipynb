# Tabular example

```python
from fastai import *          # Quick accesss to most common functionality
from fastai.tabular import *  # Quick accesss to tabular functionality     # Access to example data provided with fastai
```

Tabular data should be in a Pandas `DataFrame`.

```python
path = Path("~/.fastai/data")
df = pd.read_csv(path/'sample_data.csv')
```

```python
# Need a dummy categorical feature for the fastai APIs to work properly
df['cat_1'] = True
```

```python
df.head()
```

```python
dep_var = 'ue_loc_1'
cat_names = ['cat_1']
cont_names = ['srs_1', 'srs_2', 'srs_3', 'srs_4', 'srs_5', 'srs_6', 'srs_7', 'srs_8', 'srs_9', 'srs_10', 'srs_11', 'srs_12', 'srs_13', 'srs_14', 'srs_15', 'srs_16', 'srs_17', 'srs_18', 'srs_19', 'srs_20', 'srs_21', 'srs_22', 'srs_23', 'srs_24', 'srs_25', 'srs_26', 'srs_27', 'srs_28', 'srs_29', 'srs_30', 'srs_31', 'srs_32']
procs = [FillMissing, Categorify, Normalize]
```

```python
#test = TabularList.from_df(df.iloc[800:1000].copy(), path=path, cont_names=cont_names)
```

```python
data = (TabularList.from_df(df, path=path, cat_names=cat_names, cont_names=cont_names, procs=procs)
                           .random_split_by_pct(valid_pct=0.2, seed=42)
                           .label_from_df(cols=dep_var)
                           #.add_test(test, label=0)
                           .databunch())
```

```python
data.show_batch(rows=10)
```

```python
learn = tabular_learner(data, layers=[200,100], metrics=accuracy)
learn.fit(1, 1e-2)
```
